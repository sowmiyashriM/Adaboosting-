```python
#Cabin
df['Cabin'] = df['Cabin'].fillna('Missing')
df['Cabin'] = df['Cabin'].str[0]
df['Cabin'].value_counts()
```

```python
#dropping unwanted columns
dff = df.drop(['Name', 'Ticket'], axis = 1)
dff.head()
```

```python
dff.isna().sum()
```

```python
all_dummies = pd.get_dummies(dff, drop_first = True)
all_dummies.head()
```

```python
train = all_dummies[all_dummies['Survived'].notna()]
train.info()
```

```python
train.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train.drop(['PassengerId','Survived'],axis=1),
                                                    train['Survived'], test_size=0.30,
                                                    random_state=101, stratify = train['Survived'])
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
ada = AdaBoostClassifier(DecisionTreeClassifier(),n_estimators=100, random_state=0)
ada.fit(X_train,y_train)
```

```python
predictions = ada.predict(X_test)
```

```python
print (f'Train Accuracy - : {ada.score(X_train,y_train):.3f}')
print (f'Test Accuracy - : {ada.score(X_test,y_test):.3f}')
```
